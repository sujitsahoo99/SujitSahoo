```
def count_case_characters(input_string):
    uppercase_count = 0
    lowercase_count = 0

    for char in input_string:
        if 'A' <= char <= 'Z':
            uppercase_count += 1
        elif 'a' <= char <= 'z':
            lowercase_count += 1

    return uppercase_count, lowercase_count


user_string = input("Enter a string: ")

upper_count, lower_count = count_case_characters(user_string)

print(f"Number of uppercase characters: {upper_count}")
print(f"Number of lowercase characters: {lower_count}")
```

```
balance = 0
transactions = []

PASSWORD = "sipunzx"
MAX_ATTEMPTS = 3


def check_password():
    attempts = MAX_ATTEMPTS
    while attempts > 0:
        user_pass = input("Enter your password: ")
        if user_pass == PASSWORD:
            print("Login successful!")
            return True
        else:
            attempts -= 1
            print("Wrong password attempts left:", attempts)

    print("Too many wrong attempts")
    return False


def deposit():
    global balance
    amt = int(input("Enter Deposit Amount: "))
    balance += amt
    transactions.append("Deposited: " + str(amt))
    print("Amount Deposited")


def withdraw():
    global balance
    amt = int(input("Enter Withdrawal Amount: "))
    if amt <= balance:
        balance -= amt
        transactions.append("Withdrawn: " + str(amt))
        print("Amount Withdrawn")
    else:
        print("Insufficient Balance")


def check_balance():
    print("Current Balance:", balance)


def show_transactions():
    print("\nTransaction History")
    if len(transactions) == 0:
        print("No Transactions Yet")
    else:
        for t in transactions:
            print(t)


def menu():
    while True:
        print("\n1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Transaction History")
        print("5. Exit")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            deposit()
        elif choice == 2:
            withdraw()
        elif choice == 3:
            check_balance()
        elif choice == 4:
            show_transactions()
        elif choice == 5:
            print("Thank You")
            break
        else:
            print("Invalid Choice")


# Main Program
if check_password():
    menu()
```

```
# Data for multiple accounts
accounts = {
    "1001": {"password": "sipunzx", "balance": 0, "transactions": []},
    "1002": {"password": "abc123", "balance": 5000, "transactions": []},
}

MAX_ATTEMPTS = 3


def check_password(acc_no):
    attempts = MAX_ATTEMPTS
    while attempts > 0:
        user_pass = input("Enter your password: ")
        if user_pass == accounts[acc_no]["password"]:
            print("Login successful!")
            return True
        else:
            attempts -= 1
            print("Wrong password. Attempts left:", attempts)

    print("Too many wrong attempts")
    return False


def deposit(acc_no):
    try:
        amt = int(input("Enter Deposit Amount: "))
    except ValueError:
        print("Enter a valid number")
        return

    if amt <= 0:
        print("Amount must be positive")
        return

    accounts[acc_no]["balance"] += amt
    accounts[acc_no]["transactions"].append("Deposited: " + str(amt))
    print("Amount Deposited")


def withdraw(acc_no):
    try:
        amt = int(input("Enter Withdrawal Amount: "))
    except ValueError:
        print("Enter a valid number")
        return

    if amt <= 0:
        print("Amount must be positive")
        return

    if amt <= accounts[acc_no]["balance"]:
        accounts[acc_no]["balance"] -= amt
        accounts[acc_no]["transactions"].append("Withdrawn: " + str(amt))
        print("Amount Withdrawn")
    else:
        print("Insufficient Balance")


def check_balance(acc_no):
    print("Current Balance:", accounts[acc_no]["balance"])


def show_transactions(acc_no):
    print("\nTransaction History for Account", acc_no)
    if len(accounts[acc_no]["transactions"]) == 0:
        print("No Transactions Yet")
    else:
        for t in accounts[acc_no]["transactions"]:
            print(t)


def menu(acc_no):
    while True:
        print("\nAccount:", acc_no)
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Transaction History")
        print("5. Logout")

        try:
            choice = int(input("Enter your choice: "))
        except ValueError:
            print("Enter a valid choice")
            continue

        if choice == 1:
            deposit(acc_no)
        elif choice == 2:
            withdraw(acc_no)
        elif choice == 3:
            check_balance(acc_no)
        elif choice == 4:
            show_transactions(acc_no)
        elif choice == 5:
            print("Logging out...")
            break
        else:
            print("Invalid Choice")


# Main Program: choose account first, then login, then menu
while True:
    acc_no = input("\nEnter Account Number (or 'q' to quit): ")
    if acc_no.lower() == 'q':
        print("Goodbye")
        break

    if acc_no not in accounts:
        print("Account not found")
        continue

    if check_password(acc_no):
        menu(acc_no)
```

```
accounts = {}

def create_account():
    acc_no = input("Enter New Account NumberðŸ‘‰ðŸ»: ")

    if acc_no in accounts:
        print("Account already existsðŸ˜Ž!")
        return

    name = input("Enter Account Holder NameðŸ‘‰ðŸ»: ")
    acc_type = input("Enter Account Type (Saving/Zero)ðŸ‘‰ðŸ»: ")
    phone = input("Enter Phone NumberðŸ‘‰ðŸ»: ")
    balance = float(input("Enter Initial Deposit AmountðŸ‘‰ðŸ»: "))

    pin = input("Generate 4-digit ATM PINðŸ‘‰ðŸ»: ")

    accounts[acc_no] = {
        "name": name,
        "type": acc_type,
        "phone": phone,
        "balance": balance,
        "pin": pin,
        "history": ["Account created with balanceðŸ˜Ž" + str(balance)]
    }

    print("Account Created SuccessfullyðŸ™‚â€!")

def login():
    acc_no = input("Enter Account NumberðŸ‘‰ðŸ»: ")
    pin = input("Enter ATM PINðŸ‘‰ðŸ»: ")

    if acc_no in accounts and accounts[acc_no]["pin"] == pin:
        print("Login SuccessfulðŸ™‚â€!")
        atm_menu(acc_no)
    else:
        print("Invalid Account Number or PINðŸ˜–")

def atm_menu(acc_no):
    while True:
        print("\n---ðŸ§ATM MENUðŸ§---")
        print("1. ðŸ’µBalance Check")
        print("2. ðŸ’³Debit")
        print("3. ðŸ’³Credit")
        print("4. ðŸ“²Change Phone Number")
        print("5. ðŸ”Change ATM PIN")
        print("6. ðŸ’µTransaction History")
        print("7. ðŸšªExit")

        choice = input("Choose OptionðŸ‘‰ðŸ»: ")

        if choice == "1":
            balance_check(acc_no)
        elif choice == "2":
            debit(acc_no)
        elif choice == "3":
            credit(acc_no)
        elif choice == "4":
            change_phone(acc_no)
        elif choice == "5":
            change_pin(acc_no)
        elif choice == "6":
            transaction_history(acc_no)
        elif choice == "7":
            break
        else:
            print("Invalid OptionðŸ˜–")

def balance_check(acc_no):
    print("Available BalanceðŸ‘‰ðŸ»:", accounts[acc_no]["balance"])


def debit(acc_no):
    amount = float(input("Enter Amount to DebitðŸ‘‰ðŸ»: "))

    if amount <= accounts[acc_no]["balance"]:
        accounts[acc_no]["balance"] -= amount
        accounts[acc_no]["history"].append("Debited: " + str(amount))
        print("Amount Debited SuccessfullyðŸ˜Ž")
    else:
        print("Insufficient BalanceðŸ˜–")

def credit(acc_no):
    amount = float(input("Enter Amount to CreditðŸ‘‰ðŸ»: "))

    accounts[acc_no]["balance"] += amount
    accounts[acc_no]["history"].append("Credited: " + str(amount))
    print("Amount Credited SuccessfullyðŸ˜Ž")

def change_phone(acc_no):
    new_phone = input("Enter New Phone NumberðŸ‘‰ðŸ»: ")
    accounts[acc_no]["phone"] = new_phone
    accounts[acc_no]["history"].append("Phone Number Changed")
    print("Phone Number Updated SuccessfullyðŸ˜Ž")

def change_pin(acc_no):
    new_pin = input("Enter New 4-digit PINðŸ‘‰ðŸ»: ")
    accounts[acc_no]["pin"] = new_pin
    accounts[acc_no]["history"].append("ATM PIN Changed")
    print("ATM PIN Updated SuccessfullyðŸ˜Ž")

def transaction_history(acc_no):
    print("\n---ðŸ’µTransaction HistoryðŸ’µ---")
    for record in accounts[acc_no]["history"]:
        print(record)

def main():
    while True:
        print("\n==== ATM SYSTEM ====")
        print("1.ðŸ™‚â€Create Account")
        print("2.ðŸ”Login to ATM")
        print("3. Exit")

        option = input("Choose OptionðŸ‘‰ðŸ»: ")

        if option == "1":
            create_account()
        elif option == "2":
            login()
        elif option == "3":
            print("Thank You for visiting | see you againðŸ˜Ž!")
            break
        else:
            print("Invalid ChoiceðŸ˜–")


main()
```
